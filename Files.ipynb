```python
#Now lets set the mode as w,which means write...It willhgive error as we have open the file in write mode however we are 
#reading teh file,it has to be write when open the file in w mode
with open('myfile.txt',mode='w') as myfile:
    contents=myfile.read()
```

mode='r' is read only
mode='w' is write only(will override or create new)
mode='a' is append only will add to this files
mode='r+' is reading and writing
mode='w+' is writing and reading (Overwrite an existing file or create a new file!)
